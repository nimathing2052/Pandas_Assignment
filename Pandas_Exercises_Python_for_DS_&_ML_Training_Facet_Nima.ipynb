```python
# DO NOT CHANGE THE CODE IN THIS CELL
area = pd.Series({'California': 424,
                  'Texas': 696,
                  'New York': 141,
                  'Florida': 170,
                  'Illinois': 150
                  }) # values in thousand sq miles

population = pd.Series({'California': 38,
                  'Texas': 26,
                  'New York': 19,
                  'Florida': 20,
                  'Illinois': 129
                  }) # values in million

data = pd.DataFrame({'area': area, 'population': population}) 
data
```

Given the above DataFrame `data`, multiply each value in 'area' column by 1000, and each value in 'population' by 1000000 (1 million).

The result should look exactly like the following.

image.png

```python
# Write your code below by replacing None
data['area'] = data['area'] * 1000
data['population'] = data['population'] *1000000

data # do not change
```

```python
# Explain your code below as comments
# Correspoding columns of data is multiplied with the corresponding values given above
```

```python
# Write your code below by replacing None
data['density'] = data['population']/data['area']
data # do not change
```

```python
# Explain your code below as comments
population column of data is divided by the corresponding area to calculate the denisty
```

Sort `data` by `density` in descending order, and show top 3 states by population density.

Given the above dataframe `population`, select only the rows with 'total' (excluding 'under18') in the 'ages' column.

Follow the following steps.

1. Group the `total_population` dataframe by 'state/region', 
2. select just the 'population' column, and
3. find the mean of 'population' in each group.
4. Convert the data type of the values in the result to `int` data type.
5. Sort the values in descending order
6. Show only the first 10 values.

You will need to perform all the above steps in one line of code. 

The format will be like: `total_population`.[step1].[step2].[step3].[step4].[step5].[step6]

The result should be like the following.

image.png

```python
# Write your code below by replacing None
population_mean_by_state = total_population.groupby('state/region')['population'].mean().sort_values(ascending=False).astype(int)
population_mean_by_state.head(10) # do not change
```

```python
# Explain your code below as comments
# first groupby  with respect to 'state/region' is created and 'population' column is slected. Later, mean() function is used to find the mean and sorting is done . At las the data type is conveted to int
```

```python
# Write your code below by replacing None
high_annual_means =  msft['High'].resample('BY').mean()
high_annual_means
```
